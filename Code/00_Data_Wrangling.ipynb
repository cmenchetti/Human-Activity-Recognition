```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Training set
X_train = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\train\X_train.txt", header=None, delim_whitespace=True)
y_train = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\train\y_train.txt", header=None, delim_whitespace=True)

# Test set
X_test = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\test\X_test.txt", header=None, delim_whitespace=True)
y_test = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\test\y_test.txt", header=None, delim_whitespace=True)

# Variables labels
features = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\features.txt", header=None, sep = " ")

# subjects names
subject_test = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\test\subject_test.txt", header=None)
subject_train = pd.read_csv(r".\data\UCI HAR Dataset\UCI HAR Dataset\train\subject_train.txt", header=None)
```

#### RINOMINA VARIABILI

```python
# Lista con nomi delle variabili
features.rename({1: 'Features'}, axis=1, inplace=True)
features_list = list(features['Features'].values)
```

```python
# Rinomina variabili per X_train
for i in range(len(features_list)):
    X_train.rename({i: str(features_list[i])}, axis=1, inplace=True)

# Rinomina variabili per X_test
for i in range(len(features_list)):
    X_test.rename({i: str(features_list[i])}, axis=1, inplace=True)
    
# Rinomina variabile target y_train
y_train.rename({0: 'Label'}, axis=1, inplace=True)

# Rinomina variabile target y_test
y_test.rename({0: 'Label'}, axis=1, inplace=True)
```

#### AGGIUNTA DI COLONNA SUBJECT

In totale ci sono 30 subjects. 21 per il training set e 9 per il test set.

```python
# aggiungere subject al train
X_train['Subject'] = subject_train[0]

# aggiungere subject al test
X_test['Subject'] = subject_test[0]
```
